```python
%matplotlib notebook 

#
# let's capture this open-and-convert-to-rgb into a function
#
def opencv_open_into_rgb( image_file_name ):
    """ open image_file_name and convert to rgb """
    image_raw = cv2.imread(image_file_name, cv2.IMREAD_COLOR)  # reads into BGR
    orig_num_rows, orig_num_cols, _ = image_raw.shape          # cool underscore variable!
    num_rows, num_cols, _ = image_raw.shape
    print(f"image with name {image_file_name} read with dimensions: {image_raw.shape}")
    #
    # let's resize as long as any dimension is > 840...
    #
    maxdim = max(num_rows,num_cols)
    while maxdim > 840:
        num_rows = num_rows//2  # halve!
        num_cols = num_cols//2  # halve!
        maxdim = max(num_rows,num_cols) # try again...
        
    if orig_num_rows != num_rows: # did we resize?
        print(f"resizing to rows, cols = {num_rows}, {num_cols}")
        image_raw = cv2.resize(image_raw, dsize=(num_cols,num_rows), interpolation=cv2.INTER_LINEAR)  # resizing
        
    # now, convert to rgb
    image_rgb = cv2.cvtColor(image_raw, cv2.COLOR_BGR2RGB)     # convert from BGR to RGB
    return image_rgb

image_rgb = opencv_open_into_rgb("flag.jpg")  # "coffee.jpg"  "flag.jpg"  "big.png"

#
# let's show off how to create multiple images in one figure, too
#
fig, ax = plt.subplots(2,2)  # this means ax will be a 2x2 numpy array of axes!
ax[0,0].imshow(image_rgb)
ax[0,1].imshow(image_rgb)
ax[1,0].imshow(image_rgb)
ax[1,1].imshow(image_rgb)
```

```python
#
# nice to be able to save to a file...    and not worry about its BGR-ness
#

def save_rgb_image( image_rgb, new_file_name ):
    """ saves the image  image_rgb  to the (string) name, new_file_name
        add the extension you'd like, png, jpg, ... it knows many (not all)
    """
    image_bgr = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2BGR)     # convert from BGR to RGB
    result = cv2.imwrite(new_file_name,image_bgr)
    if result == True:
        print(f"image_rgb was saved to {new_file_name}")
    else:
        print(f"there was a problem saving image_rgb to {new_file_name}")
    
#
# try it!
#
save_rgb_image( image_rgb, "new_image.png" )  # it needs the extension
```

```python
def one_image_filter_rgb_example( image_rgb ):
    """ an example of a pixel-by-pixel filter 
            this one keeps ONLY the green channel, setting the r and b channels to 0
            old pixel value:   [42,42,100]   
            new pixel value:   [ 0,42,  0]
        input: an r, g, b image
        output: a transformed r, g, b image
    """
    image_new = image_rgb.copy() # start with a copy of the original...
    num_rows, num_cols, num_chans = image_rgb.shape
    
    for row in range(num_rows):  # loop over each row
        for col in range(num_cols):  # loop over each column
            r, g, b = image_rgb[row,col]   # the "old" pixel's rgb
            #
            # remember that the pixels need to be integers from 0 to 255 (inclusive)
            #
            image_new[row,col] = 0, g, 0   # let's try some others!  (move g to the first or last channel?)
            
    return image_new

image_rgb = opencv_open_into_rgb("flag.jpg")  # "coffee.jpg" # try others!
image_new = one_image_filter_rgb_example(image_rgb)
fig, ax = plt.subplots(2,1)  # this means ax will be a 2x1 numpy array (it's 1d!)
ax[0].imshow(image_rgb)
ax[1].imshow(image_new)
```

```python
def one_image_filter_rgb_example2( image_rgb ):
    """ an example of a pixel-by-pixel filter 
            this one sets to "max green" any pixel with a '1' (a one digit) in its (decimal) green value
            old green value:   42   # this is _not_ set to green
            new green value:   142  # this _is_ set to green [0,255,0]
        input: an r, g, b image
        output: a transformed r, g, b image
    """
    image_new = image_rgb.copy() # start with a copy of the original...
    num_rows, num_cols, num_chans = image_rgb.shape
    
    for row in range(num_rows):  # loop over each row
        for col in range(num_cols):  # loop over each column
            r, g, b = image_rgb[row,col]   # the "old" pixel's rgb
            
            # remember:  we can do anything with the pixel data... here, let's check if there's a '1' in g:
            green_as_str = str(g)
            if '1' in green_as_str:  # not in photoshop! :-) 
                image_new[row,col] = [0,255,0]
            else:
                image_new[row,col] = [0,0,0] # or [0,0,0] or [255,255,255] or...       
            #
            # remember that the pixels need to be integers from 0 to 255 (inclusive)
            #
            
    return image_new

image_rgb = opencv_open_into_rgb("flag.jpg")  # "coffee.jpg" # try others!
image_new = one_image_filter_rgb_example2(image_rgb)
fig, ax = plt.subplots(2,1)  # this means ax will be a 2x1 numpy array (it's 1d!)
ax[0].imshow(image_rgb)
ax[1].imshow(image_new)
```

```python
def bin2num(s):
    """ convert from binary (a string, s) to a base-10 int, the number n """
    if s == '': return 0
    return 2*bin2num(s[:-1]) + int(s[-1])

def num2bin(n):
    """ convert from a number (an int n) to a string in binary, s """
    if n == 0: return ''
    return num2bin(n//2) + str(n%2)

def one_image_filter_rgb_example3( image_rgb ):
    """ an example of a pixel-by-pixel filter:  
            this one sets the "bottom two bits" of each green value to zeros, e.g.,
            old green value:   '00101010'   # always 8 bits...
            new green value:   '00101000'   # always 8 bits, now ending in two zeros
        input: an r, g, b image
        output: a transformed r, g, b image
    """
    image_new = image_rgb.copy() # start with a copy of the original...
    num_rows, num_cols, num_chans = image_rgb.shape
    
    for row in range(num_rows):  # loop over each row
        for col in range(num_cols):  # loop over each column
            r, g, b = image_rgb[row,col]   # the "old" pixel's rgb
            
            # remember:  we can do anything with the pixel data... here, let's convert to binary
            green_in_binary = num2bin(g)
            nbits = len(green_in_binary)  # to make sure we have 8 bits...
            green_in_binary = '0'*(8-nbits) + green_in_binary
            if row%200 == 42 and col%200 == 42: print(f"\nold green_in_binary: {green_in_binary} at {row},{col}")
            
            # now, let's make the last two bits equal to zero...
#             green_in_binary = green_in_binary[:-2] + '00'  # done!
#             if row%200 == 42 and col%200 == 42: print(f"new green_in_binary: {green_in_binary} at {row},{col}\n")
            
            new_g = bin2num(green_in_binary)  # convert back to base 10
            
            #
            # remember that the pixels need to be integers from 0 to 255 (inclusive)
            #
            image_new[row,col] = r, new_g, b
            
    return image_new

image_rgb = opencv_open_into_rgb("flag.jpg")  # "coffee.jpg" # try others!
image_new = one_image_filter_rgb_example3(image_rgb)
fig, ax = plt.subplots(2,1)  # this means ax will be a 2x1 numpy array (it's 1d!)
ax[0].imshow(image_rgb)
ax[1].imshow(image_new)
```

```python
def two_image_filter_rgb( image_rgb1, image_rgb2 ):
    """ an example of a two-image pixel-by-pixel filter
            this takes the componentwise min! hmmm...
        input: two r, g, b images
        output: a transformed r, g, b image (the "min image")
    """
    num_rows1, num_cols1, num_chans1 = image_rgb1.shape
    num_rows2, num_cols2, num_chans2 = image_rgb2.shape
    
    num_rows = min(num_rows1,num_rows2)
    num_cols = min(num_cols1,num_cols2)
    # create a blank image
    image_new = np.zeros((num_rows,num_cols,3), np.uint8)  # we're assumeing 3 channels
    
    for row in range(num_rows):  # loop over each row
        for col in range(num_cols):  # loop over each column
            r1, g1, b1 = image_rgb1[row,col]   # the "old" pixel1's rgb
            r2, g2, b2 = image_rgb2[row,col]   # the "old" pixel2's rgb

            r = min(r1, r2)
            g = min(g1, g2)
            b = min(b1, b2)
            
            image_new[row,col] = [r,g,b]

            #
            # remember that the pixels need to be integers from 0 to 255 (inclusive)
            #
            
    return image_new
    
    

image_rgb1 = opencv_open_into_rgb("flag.jpg")  # "coffee.jpg" # try others!
image_rgb2 = opencv_open_into_rgb("coffee.jpg")  # "flag.jpg" # try others!
image_new = two_image_filter_rgb(image_rgb1,image_rgb2)
fig, ax = plt.subplots(3,1)  # this means ax will be a 3x1 numpy array (it's 1d!)
ax[0].imshow(image_rgb1)
ax[1].imshow(image_rgb2)
ax[2].imshow(image_new)
```

```python
def two_image_filter_rgb_overlap( image_rgb1, image_rgb2, ulx, uly ):
    """ an example of a two-image pixel-by-pixel filter
            puts im2 on top of im1 at upper-left-x (ulx) (col) and upper-left-y (uly) (row)
            the resulting size is image_rgb1 (the rest of image_rgb2 is dropped...)
        input: two r, g, b images
        output: a transformed r, g, b image (the "min image")
    """
    num_rows1, num_cols1, num_chans1 = image_rgb1.shape
    num_rows2, num_cols2, num_chans2 = image_rgb2.shape
    
    image_new = image_rgb1.copy()   # not symmetric! image_rgb1 is the "background"
    
    max_row = min(num_rows1,uly+num_rows2) # no more than num_rows1
    max_col = min(num_cols1,ulx+num_cols2) # no more than num_cols1
    
    for row in range(uly,max_row):
        for col in range(ulx,max_col):
            image_new[row,col] = image_rgb2[row-uly,col-ulx]  # put 2nd image over 1st
            
    return image_new



image_rgb1 = opencv_open_into_rgb("flag.jpg")  # "coffee.jpg" # try others!
image_rgb2 = opencv_open_into_rgb("coffee.jpg")  # "flag.jpg" # try others!
image_new = two_image_filter_rgb_overlap(image_rgb1,image_rgb2,42,42)
#image_new = two_image_filter_rgb_overlap(image_rgb2,image_rgb1,42,42)  # not symmetric

fig, ax = plt.subplots(3,1)  # this means ax will be a 3x1 numpy array (it's 1d!)
ax[0].imshow(image_rgb1)
ax[1].imshow(image_rgb2)
ax[2].imshow(image_new)
```
